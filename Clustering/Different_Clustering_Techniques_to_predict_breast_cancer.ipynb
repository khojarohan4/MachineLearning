```python
# import the 'Breast Cancer' dataset from sklearn as below
from sklearn import datasets 
data = datasets.load_breast_cancer()
```

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
x_data = pd.DataFrame(data.data,columns = data.feature_names)
```

```python
y_data = pd.DataFrame(data.target,columns = ['benign/malignant'])
```

```python
y_data = data.target
```

```python
# Create 2 clusters for the dataset using KMeans clustering
from sklearn.cluster import KMeans
km = KMeans(n_clusters=2, init="k-means++", n_init=10)
km_pred = km.fit_predict(x_data)
```

```python
km_pred
```

```python
y_data.head()
```

```python
# visualise the KMeans clusters
import matplotlib.pyplot as plt
%matplotlib inline
# colors = ['green','red']
plt.figure(figsize = (15, 10))

f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)

ax1.scatter(x_data.iloc[:,0], x_data.iloc[:,1], c=y_data["benign/malignant"], cmap="jet", edgecolor="None", alpha=0.35)
ax1.set_title("Actual clusters")

ax2.scatter(x_data.iloc[:,0], x_data.iloc[:,1], c=km_pred, cmap="jet", edgecolor="None", alpha=0.35)
ax2.set_title("KMeans clustering plot")
```

```python
# Create Hierarchicial clustering model using linkages : {“ward”, “complete”, “average”}
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=2, linkage="ward")
ac_pred = ac.fit_predict(x_data)
```

```python
#visualize the clusters
plt.figure(figsize = (15, 10))
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)

ax1.scatter(x_data.iloc[:,0], x_data.iloc[:,1], c=y_data["benign/malignant"], cmap="jet", edgecolor="None", alpha=0.35)
ax1.set_title("Actual clusters")

ax2.scatter(x_data.iloc[:,0], x_data.iloc[:,1], c=ac_pred, cmap="jet", edgecolor="None", alpha=0.35)
ax2.set_title("Agglomeratve clustering plot with ward linkage")
```

```python
from sklearn.cluster import AgglomerativeClustering
ac1 = AgglomerativeClustering(n_clusters=2, linkage="complete")
ac1_pred = ac1.fit_predict(x_data)
plt.figure(figsize = (15, 10))
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)

ax1.scatter(x_data.iloc[:,0], x_data.iloc[:,1], c=y_data["benign/malignant"], cmap="jet", edgecolor="None", alpha=0.35)
ax1.set_title("Actual clusters")

ax2.scatter(x_data.iloc[:,0], x_data.iloc[:,1], c=ac1_pred, cmap="jet", edgecolor="None", alpha=0.35)
ax2.set_title("Agglomeratve clustering plot with complete linkage")
```
